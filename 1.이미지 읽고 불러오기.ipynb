## opencv를 이용한 방법

1. 사용할 이미지 파일을 model.png파일로 우리가 작성할 코드샘플 폴더 하위에 image에 저장해서 사용할 것이다. 이를 사용하기 위해서 먼저 import cv2와 numpy가 필요하다.

```python
import cv2
import numpy as np
```

```python
import numpy as np
import cv2

def showImage():
    imgfile = 'image/lenna.png'
    img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
    
    cv2.imshow('lenna', img)
    cv2.waitKey(0)#키보드 누를 때 까지 대기하는 코드
    cv2.destroyAllWindows()
    
showImage()
```

이미지를 읽고 화면에 디스플레이하는 함수를 만들었다. 이는 showImage()

img = cv2.imread(imgfile, cv2.IMREAD_COLOR)

imread는 이미지파일을 읽기 위한 객채를 리턴한다. 첫번째는 읽고자 하는 파일경로 두번째는 이미지 파일을 읽는 방식이다. 방식에는 3가지 방법이 존재합니다.

- cv.IMREAD_COLOR: 컬러이미지, 이미지의 투명한부분은 무시됨, 디폴트 플래그 정수값은 1
- cv.IMREAD_GRAYSCALE: 흑백 이미지로 로드 정수값은 0
- cv.IMREAD_UNCHANGED: 알파채널을 포함하여 이미지 그대로 로드함 정수값은 -1

cv.imshow('model', img)

반환된 이미지를 화면에 나타내기 위한 함수이다. 첫번째는 타이틀 두번째는 불러올 파일

2. 다음은 이미지 크기를 변경시킬 수 있는 창을 만들기 위해서 cv2.namedWindow()함수를 이용해서 창의 성격을 지정할 수 있습니다.

```python
import numpy as np
import cv2

def showImage():
    imgfile = 'image/lenna.png'
    img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
    
    cv2.namedWindow('lenna',cv2.WINDOW_NORMAL)
    cv2.imshow('lenna', img)
    cv2.waitKey(0)#키보드 누를 때 까지 대기하는 코드
    cv2.destroyAllWindows()
    
showImage()
```

다음 함수는 두가지가 존재할 수 있다.

- cv2.WINDOW_AUTOSIZE : 원본 이미지 크기로 고정하여 생성
- cv2.WINDOW_NORMAL : 원본이미지로 생성한뒤 사용자가 크기조정가능

단, 주의점은 cv2.imshow()에서 지정한타이틀과 같아야함

3. 이미지파일의 복사본을 만드는 것

```python
import numpy as np
import cv2

def showImage():
    imgfile = 'image/lenna.png'
    img = cv2.imread(imgfile, cv2.IMREAD_COLOR)
    cv2.imshow('lenna', img)
    
    k = cv2.waitKey(0) & 0xFF #키보드값을 아스키코드로 반환
    
    if k ==27:     
        cv2.destroyAllWindows() #ESC를 누를 때 
    elif k == ord('c'):#ord('c') 아스키코드'c' 값인 99 호출 반대는 chr(99) c를 누르면 복사
        cv2.imwrite('image/lenna_copy.jpg',img)#image/lenna_copy.jpg로 저장
        cv2.destroyAllWindows()
        
    
showImage()
```

## Matplotlib을 이용하여 화면에 이미지 표시

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

def showImage():
    imgfile = 'image/lenna.png'
    img = cv2.imread(imgfile, cv2.IMREAD_GRAYSCALE)
    '''
    matplotlib은 RGB모드를 사용하고 openCV는 BGR모드를 사용한다. 그래서 컬러가 재대로
    나오지 않기 때문에 grayscale을 사용
    '''
    plt.imshow(img, cmap='gray', interpolation ='bicubic')
    plt.xticks([])#기본적인
    plt.yticks([])
    plt.title('lenna')
    plt.show()
    
showImage()
```
